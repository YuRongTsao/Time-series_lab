```python
import numpy as np
from random import seed
from random import random
import statsmodels.tsa.api as smt
from statsmodels.tsa.stattools import adfuller
from random import random
import matplotlib.pyplot as plt
from pprint import pprint
from statsmodels.graphics.tsaplots import plot_acf

max_lag = 20
n = 500 # lots of samples to help estimates
```

```python
def generateARMA12Points(params):
    np.random.seed(1)
    x = a = np.random.normal(size=n) 
    
    for t in range(2,n):
        x[t] = params[0] + params[1] * x[t-1] + a[t] + params[2] * a[t-1] + params[3] * a[t-2]
        
    return x
```

```python
def drawPlot(points):
    x = range(1,len(points)+1) 
    plt.plot(x, points,color = 'b',linewidth=1,alpha=0.6,label='observation')
    plt.legend(loc='upper right')
    plt.show()
```

```python
def drawACF(points):
    plot_acf(points,lags = 30,markersize = 3 )
    plt.title('ACF')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()
```

```python
def generateRandomWalk():
    x = a = np.random.normal(size=n) 
    
    for t in range(1,n):
        x[t] = 0.3 + x[t-1] + a[t]
        
    return x
```

```python
def DickeyFuller(points):
    return adfuller(points)[1]
```

```python
def difference(points, interval=1):
    return [points[i] - points[i - interval] for i in range(interval, len(points))]
        
```
